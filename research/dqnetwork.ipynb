```python
class DQN(nn.Module):
    def __innit__(self,input_dim,output_dim):
        super().__init__()
        self.fc1 = nn.Linear(input_dim,128)
        self.fc2 = nn.Linear(128,128)
        self.fc3 = nn.Linear(128, output_dim)
        
    def forward(self, x):
        x = torch.reul(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)
    
# the size of a step when updating a weight, can overoptimize or underoptimize
learning_rate = 0.001
# the discount factor ( immediate rewards vs future rewards )
gamma = 0.99

# the exploration vs exploitation rate 
# set to max ( exploration )
epsilon = 1.0
# set to min ( exploitation )
epsilon_min = 0.01
# contols how fast the agent switches from exploration to exploitation ( stops being random and starts using what it has learned. )
epsilon_decay = 0.995

# input size
batch_size = 64
# the agent has 2 networks, policy and target ( online is the network which undergoes weight updating frequently whereas policy updates weight updating only after 1000 steps) 
target_update_freq = 1000
# For storing past experiences
memory_size = 10000

# A full game ( from start to end ). The agent will play 100 full games
episodes = 1000

# so I have the following questions in the below given peice of code
# 1) How does env.observation_space.shape work ? what is it ? Same for action_space ?
# 2) normally for calling the neural network Id use model right ? and not DQN ? 
# 3) What is policy net and target net ? 
# 4) Can you explain what the last 2 lines does ? 

# these are gymnasium variable and they are used to define the input dimensions and output dimensions for our example of cartpole
input_dim = env.observation_space.shape[0]
output_dim = env.action_space.n
# the 2 networks that this method uses 
policy_net = DQN(input_dim, output_dim)
target_net = DQN(input_dim, output_dim)

target_net.load_state_dict(policy_net.state_dict())
target_net.eval()
```
